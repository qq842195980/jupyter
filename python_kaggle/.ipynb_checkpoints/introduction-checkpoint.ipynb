# 简介篇

## 综述

机器学习系统具备以下特点：
- 无法直接使用固定规则或者流程代码完成的问题
- 具备“学习”能力，具有对未来的泛化能力（Generalization）
- 具备不断改善自身应对具体任务的能力

> *A program can be said to learn from experience __E__ with respect to some class of tasks __T__ and performance measure __P__, if its performance at tasks in __T__, as ,measured by __P__, imporoves with experience __E__ --- Tom Mitchell(Carnegie Mellon University)*

### 任务（Task）
监督学习（Supervised Learning）
- Classification
- Regression
无监督学习（Unsupervised Learning）
- Dimensionality Reduction
- Clustering

### 经验（Experience）
经验：数据中对学习任务有用的特定信息。
特征：反映数据内在规律的信息。

### 性能（Performance）
性能：评价所完成任务质量的指标。
